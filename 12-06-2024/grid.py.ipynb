```python
import matplotlib.pyplot as plt
```

```python
#data
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

#create the plot

plt.plot(x,y1,label='r1',linestyle='-',color='b',linewidth='3',marker='o',markersize=8,markerfacecolor="orange")
plt.plot(x,y2,label='r2',linestyle='--',color='r',linewidth='3',marker='o',markersize=8,markerfacecolor="yellow")

#Add a title
plt.title('Customize Lineplot with grid')

#Add the labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#create a legend method with the chart that we have created
plt.legend()

#add the value as a grid
plt.grid(True)

#customize the grid
plt.grid(color="grey",linestyle='--',linewidth=1)

plt.show()
```
