```python
import matplotlib.pyplot as plt
```

```python
#dataset
x=[1,2,3,4,5,6,7]
y=[2,3,4,5,7,11,13]
```

```python
#create a plot with the marker
plt.plot(x,y,marker='o',linestyle="-.",color="b",markersize=6,markerfacecolor="red")

#Add a title
plt.title('Line Plot with markers')

#Add the labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
