```python
import matplotlib.pyplot as plt
```

```python
#dataset
x=[1,2,3,4,5,6,7]
y=[2,3,4,5,7,11,13]

#create a plot for our data
plt.plot(x,y)

#customization for the plot
#Add a title
plt.title('Line Plot')

#Add the labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#output the plot
plt.show()
```
