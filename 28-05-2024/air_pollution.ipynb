```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = 'air-pollution.csv'
df = pd.read_csv(file_path)

# Clean the data by removing rows that are completely empty
df_cleaned = df.dropna(how="all")

# Fill the remaining NaN values with 0
df_cleaned.fillna(0, inplace=True)

# Remove duplicate entries
df_cleaned.drop_duplicates(inplace=True)

# Filter based on the country
India_data = df_cleaned[df_cleaned['Entity'] == 'India']
```

```python
pd.set_option('display.max_rows', None)
India_data
```

```python
# Calculate median, mean, and std for relevant columns and store them in a new DataFrame
stats = {
    'Statistic': ['Median', 'Mean', 'Std'],
    'NOx': [
        India_data['Nitrogen oxide (NOx)'].median(),
        India_data['Nitrogen oxide (NOx)'].mean(),
        India_data['Nitrogen oxide (NOx)'].std()
    ],
    'SO2': [
        India_data['Sulphur dioxide (SO₂) emissions'].median(),
        India_data['Sulphur dioxide (SO₂) emissions'].mean(),
        India_data['Sulphur dioxide (SO₂) emissions'].std()
    ],
    'CO': [
        India_data['Carbon monoxide (CO) emissions'].median(),
        India_data['Carbon monoxide (CO) emissions'].mean(),
        India_data['Carbon monoxide (CO) emissions'].std()
    ]
}
```

```python
# Create a new DataFrame to store the statistics
stats_df = pd.DataFrame(stats)

# Display the statistics DataFrame
print(stats_df)
```
